```python
with plt.style.context('classic'):
  plt.figure(figsize = (20, 10))
  plt.xticks(rotation=45, ha='right')
  plt.xlabel('Date')
  plt.ylabel('Number of registered users')
  plt.plot(df['date'], df['registered']);
```

**З графіку видно, що кількість клієнтів, які регеструються у додатку зростає.** Але як на мене, то у грудні активність просідає у порівнянні з іншими місяцями. Але потім активність знову зростає.Можливо це пов'язано з:

  * новорічними святами,і з довгими відпусткам на роботі, бо якщо люди більше проводять часу вдома, то вони й не користуються велосипедами.

  * або з недостатньою кількістю маркетингових кампаній, бо можливо продумати якусь новорічну акцію, щоб заохотити клієнтів.
